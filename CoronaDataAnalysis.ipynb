```python
df_choosen.sum(axis=0).plot.bar()
```

```python
sum_daily=df_choosen.cumsum(axis=0)
sum_daily
```

```python
sum_daily.plot()
```

```python
sum_daily['Turkey'].plot.area()
```

```python
sum_daily['Turkey'].plot()
```

```python
# drop date column
df_countries=df.drop(['date'],axis=1)
df['Total']=df_countries.sum(axis=1)
df['Total'].plot();
```

```python
df_countries['Total']=df_countries.sum(axis=1)
df_countries=df_countries.cumsum(axis=0)
df_countries['Total'].plot()
```

```python
df_countries[['Turkey','Italy','Germany','United Kingdom','China','US']].plot()
```

```python
df_countries['date']=df['date']
df_countries
```

```python
df_countries.set_index('date',inplace=True)
df_countries
```

```python
df_countries_choosen=df_countries[['Turkey','Italy','Germany','United Kingdom','China','US']]
df_countries_choosen.plot()
```

## Visualization with Bar Chart Race

```python
bcr.bar_chart_race(df_countries_choosen,filename='covid_19_data.mp4',figsize=(10,8),period_length=500,steps_per_period=10,title='Covid-19 @SenasuDemir')
```
